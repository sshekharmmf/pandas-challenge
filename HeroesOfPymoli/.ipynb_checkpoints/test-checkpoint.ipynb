```python
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
#Display the total number of players
players = {
'Total Players':[purchase_data["SN"].nunique()]
}
players_count = pd.DataFrame(players)
(players_count)
```

```python
# #Purchasing Analysis (Total)
# Number of Unique Items
# Average Purchase Price
# Total Number of Purchases
# Total Revenue
Total_unique_item=purchase_data["Item ID"].nunique()
#print(Total_unique_item)
Averge_price=purchase_data["Price"].mean()
Number_of_Purchases=purchase_data["Purchase ID"].count()
Total_Revenue=purchase_data["Price"].sum()
Total_Revenue


purchase_analysis = {
    '#Unique Items': [purchase_data["Item ID"].nunique()],
    'Average Price': [purchase_data["Price"].mean()],
    'Number of Purchases':[purchase_data["Purchase ID"].count()],
    'Total Revenue': [purchase_data["Price"].sum()]
}

purchase_analysis_df = pd.DataFrame(purchase_analysis)
purchase_analysis_df

```

```python
#create new datata frame by removing duplicate
unique_players_df= purchase_data.loc[:,["SN", "Gender"]]
unique_players_df2=unique_players_df.drop_duplicates()
unique_players_df2.head()
```

```python
# Calculate players by gender
players_by_gender = unique_players_df2['Gender'].value_counts()
players_by_gender_df = pd.DataFrame(players_by_gender)

percent_players = (unique_players_df2['Gender'].value_counts() / (unique_players_df2['Gender'].count())) * 100


# Add 'Percent of Players' to table
players_by_gender_df['Percent of Players'] = percent_players
gender_demographics_df = pd.DataFrame(players_by_gender_df)
gender_demographics_df


# Rename 'Gender' to 'Total Count'
gender_demographics_df2 = gender_demographics_df.rename(columns={
    'Gender': 'Total Count',
})
gender_demographics_df2
```

```python
#Purchasing Analysis (Gender)
#Purchase Count
all_gender_df = purchase_data.groupby('Gender')
purchase_count = all_gender_df['Gender'].count()
player_count_df = pd.DataFrame(purchase_count)
 #Calculate average purchase price by gender
avg_price = all_gender_df['Price'].mean()

# Calculate total purchase price by gender
total_gender_purchase = all_gender_df['Price'].sum()

#Calculate Average Purchase Total per Person by Gender
norm_total = total_gender_purchase / players_by_gender

#Add these together to the table

player_count_df['Average Purchase Price'] = avg_price
gender_purchase_df = pd.DataFrame(player_count_df)

gender_purchase_df['Total Purchase Value'] = total_gender_purchase
gender_purchase_df2 = pd.DataFrame(gender_purchase_df)

gender_purchase_df2['Avg Total Purchase Per Person'] = norm_total
gender_purchase_df3 = pd.DataFrame(gender_purchase_df)

gender_purchase_df3
```
