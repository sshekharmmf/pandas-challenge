```python
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
#Display the total number of players
players = {
'Total Players':[purchase_data["SN"].nunique()]
}
players_count = pd.DataFrame(players)
(players_count)
```

```python
# #Purchasing Analysis (Total)
# Number of Unique Items
# Average Purchase Price
# Total Number of Purchases
# Total Revenue
Total_unique_item=purchase_data["Item ID"].nunique()
#print(Total_unique_item)
Averge_price=purchase_data["Price"].mean()
Number_of_Purchases=purchase_data["Purchase ID"].count()
Total_Revenue=purchase_data["Price"].sum()
Total_Revenue


purchase_analysis = {
    '#Unique Items': [purchase_data["Item ID"].nunique()],
    'Average Price': [purchase_data["Price"].mean()],
    'Number of Purchases':[purchase_data["Purchase ID"].count()],
    'Total Revenue': [purchase_data["Price"].sum()]
}

purchase_analysis_df = pd.DataFrame(purchase_analysis)
purchase_analysis_df

```

```python
#create new datata frame by removing duplicate
unique_players_df= purchase_data.loc[:,["SN", "Gender"]]
unique_players_df2=unique_players_df.drop_duplicates()
unique_players_df2.head()
```

```python
# Calculate players by gender
players_by_gender = unique_players_df2['Gender'].value_counts()
players_by_gender_df = pd.DataFrame(players_by_gender)

percent_players = (unique_players_df2['Gender'].value_counts() / (unique_players_df2['Gender'].count())) * 100


# Add 'Percent of Players' to table
players_by_gender_df['Percent of Players'] = percent_players
gender_demographics_df = pd.DataFrame(players_by_gender_df)
gender_demographics_df


# Rename 'Gender' to 'Total Count'
gender_demographics_df2 = gender_demographics_df.rename(columns={
    'Gender': 'Total Count',
})
gender_demographics_df2
```

```python
#Purchasing Analysis (Gender)
#Purchase Count
all_gender_df = purchase_data.groupby('Gender')
purchase_count = all_gender_df['Gender'].count()
player_count_df = pd.DataFrame(purchase_count)
 #Calculate average purchase price by gender
avg_price = all_gender_df['Price'].mean()

# Calculate total purchase price by gender
total_gender_purchase = all_gender_df['Price'].sum()

#Calculate Average Purchase Total per Person by Gender
norm_total = total_gender_purchase / players_by_gender

#Add these together to the table

player_count_df['Average Purchase Price'] = avg_price
gender_purchase_df = pd.DataFrame(player_count_df)

gender_purchase_df['Total Purchase Value'] = total_gender_purchase
gender_purchase_df2 = pd.DataFrame(gender_purchase_df)

gender_purchase_df2['Avg Total Purchase Per Person'] = norm_total
gender_purchase_df3 = pd.DataFrame(gender_purchase_df)

gender_purchase_df3
```

```python
# Age Demographics
# The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
# Purchase Count
# Average Purchase Price
# Total Purchase Value
# Average Purchase Total per Person by Age Group

bins = [0, 9, 14, 19,24,29,34,39,100]
group_names=['<10','10-14','15-19','19-24','25-29','20-34','35-39','40+']
```

```python
purchase_data["age_summary"] = pd.cut(purchase_data["Age"], bins, labels=group_names)
df = purchase_data.groupby("age_summary")
total_count_age = df["Purchase ID"].count()
age_purchase_df=pd.DataFrame(total_count_age)
avg_purchase_price_age = df ["Price"].mean()
total_purchase_value = df["Price"].sum()
total_count_age = df["SN"].nunique()
avg_purchase_per_person_age = total_purchase_value/total_count_age


#Add these together to the table



age_purchase_df['Average Purchase Price'] =avg_purchase_price_age
age_purchase_df2 = pd.DataFrame(age_purchase_df)

age_purchase_df2['Total Purchase Value'] = total_purchase_value 
age_purchase_df3 = pd.DataFrame(age_purchase_df2)

age_purchase_df3['Avg Total Purchase per Person']=avg_purchase_per_person_age
age_purchase_df4= pd.DataFrame(age_purchase_df3)

(age_purchase_df4)
```

```python
#Top Spenders

df_spender = purchase_data.groupby("SN")

purchase_count_sp = df_spender["Purchase ID"].count()

purchase_count_sp_df=pd.DataFrame(purchase_count_sp)

avg_Purchase_Value = df_spender["Price"].mean()

Total_Purchase_Value = df_spender["Price"].sum()


#Add these together to the table



purchase_count_sp_df['Total purchase value'] =Total_Purchase_Value 
purchase_count_sp_df['Average Purchase Price'] =avg_Purchase_Value 
purchase_count_sp_df2 = pd.DataFrame(purchase_count_sp_df)
purchase_count_sp_df3=purchase_count_sp_df2.rename(columns={"Purchase ID": "Purchase Count"})

# Sort in descending order to obtain top 5 popular item 
top_spender= purchase_count_sp_df3.sort_values(["Purchase Count"], ascending=False).head()
top_spender
```

```python
# Identify the 5 most popular items by purchase count, then list (in a table):
# Item ID
# Item Name
# Purchase Count
# Item Price
# Total Purchase Value

items = purchase_data[["Item ID", "Item Name", "Price"]]

df_item = items.groupby(["Item Name","Item ID"])

purchase_count_item = df_item["Item ID"].count()

purchase_count_item_df=pd.DataFrame(purchase_count_item)

Total_Purchase_Value = df_item["Price"].sum()

# Find individual item price
item_price = Total_Purchase_Value/purchase_count_item


#Add these together to the table



purchase_count_item_df['Total purchase value'] =Total_Purchase_Value 
purchase_count_item_df['Item Price'] =item_price
purchase_count_item_df2 = pd.DataFrame(purchase_count_item_df)
purchase_count_item_df3=purchase_count_item_df2.rename(columns={"Item ID": "Purchase Count"})

# Sort in descending order to obtain top 5 popular item 
popular_item = purchase_count_item_df3.sort_values(["Purchase Count"], ascending=False).head()


```

```python
#Identify the 5 most profitable items by total purchase value, then list (in a table):

items = purchase_data[["Item ID", "Item Name", "Price"]]

df_item = items.groupby(["Item Name","Item ID"])

purchase_count_item = df_item["Item ID"].count()

purchase_count_item_df=pd.DataFrame(purchase_count_item)

Total_Purchase_Value = df_item["Price"].sum()

# Find individual item price
item_price = Total_Purchase_Value/purchase_count_item


#Add these together to the table



purchase_count_item_df['Total purchase value'] =Total_Purchase_Value 
purchase_count_item_df['Item Price'] =item_price
purchase_count_item_df2 = pd.DataFrame(purchase_count_item_df)
purchase_count_item_df3=purchase_count_item_df2.rename(columns={"Item ID": "Purchase Count"})

# Sort in descending order to obtain top 5 popular item 
proftable_item = purchase_count_item_df3.sort_values(["Total purchase value"], ascending=False).head()

```
